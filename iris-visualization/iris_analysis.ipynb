```python
import matplotlib.pyplot as plt
import seaborn as sns
iris=sns.load_dataset("iris")
```

```python

print(iris)
```

```python
plt.figure(figsize=(10, 4))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(iris['sepal_length'], color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Sepal Length")

# Horizontal Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=iris['sepal_length'], color='lightgreen')
plt.title("Boxplot")
plt.xlabel("Sepal Length")

plt.tight_layout()
plt.show()
```
